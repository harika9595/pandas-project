```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Simulate demographics data
demographics = pd.DataFrame({
    'patient_id': [f'P{i:03d}' for i in range(1, 21)],
    'age': np.random.randint(30, 70, size=20),
    'gender': np.random.choice(['M', 'F'], size=20),
    'group': np.random.choice(['treatment', 'control'], size=20)
})

# Simulate lab results
lab_results = pd.DataFrame({
    'patient_id': np.random.choice(demographics['patient_id'], size=18, replace=False),
    'ALT': np.random.randint(15, 100, size=18),
    'AST': np.random.randint(15, 100, size=18)
})
```

```python
merged = pd.merge(demographics, lab_results, on='patient_id', how='left')
print(merged.head())
```

```python
abnormal_alt = merged[merged['ALT'] > 50]
print(abnormal_alt[['patient_id', 'ALT', 'group']])
```

```python
avg_by_group = merged.groupby('group')[['ALT', 'AST']].mean().reset_index()

avg_by_group.plot(x='group', kind='bar', figsize=(8, 5), title='Average Liver Enzymes by Group')
plt.ylabel('Enzyme Level')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.tight_layout()
plt.show() 

```
